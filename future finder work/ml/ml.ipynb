```python
df= pd.read_csv(r"C:\Users\darsh_3lhmtua\pythone start\future finder work\csv\insurance_data.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[["age"]],df["bought_insurance"],test_size=0.1)
```

```python
plt.scatter(df["age"],df["bought_insurance"])
```

```python
model= LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
y_test
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_test,y_predict)
```

```python
model.score(x_test,y_predict)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_predict,y_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt

X_range = pd.DataFrame({"age": np.linspace(df["age"].min(), df["age"].max(), 300)})
y_prob = model.predict_proba(X_range)[:, 1]

plt.figure(figsize=(10, 6))
plt.scatter(df["age"], df["bought_insurance"], color="red", label="Actual data")
plt.plot(X_range["age"], y_prob, color="blue", label="Logistic regression curve")
plt.axhline(0.5, color="gray", linestyle="--", label="Threshold=0.5")
plt.title("Logistic Regression - Sigmoid Curve")
plt.xlabel("Age")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

```python
df= pd.read_csv(r"C:\Users\darsh_3lhmtua\pythone start\future finder work\csv\Salaries.csv")
```
